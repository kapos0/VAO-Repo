```python
ad_soyad = input("Lütfen ad soyad giriniz: ")
yas = input("Lütfen yaşınızı giriniz: ")
sehir = input("Lütfen yaşadığınız şehri giriniz: ")

print(f"--- Profil Kartı ---", end="\n\n")
print("Ad Soyad:", ad_soyad, sep=" -> ")
print(f"Yaş: {yas}")
print(f"Şehir: {sehir}")
```

```python
harcamalar = [120, 45, 230, 15, 80, 60]
toplam_harcama = 0
yuz_ustu_sayisi = 0

for harcama in harcamalar:
    toplam_harcama += harcama
    if harcama > 100:
        yuz_ustu_sayisi += 1

en_buyuk = max(harcamalar)
en_kucuk = min(harcamalar)

print(f"Toplam Harcama: {toplam_harcama} TL")
print(f"En Büyük: {en_buyuk} TL, En Küçük: {en_kucuk} TL")
print(f"100 TL üzeri harcama adedi: {yuz_ustu_sayisi}")

if toplam_harcama > 500:
    indirimli_toplam = toplam_harcama * 0.90
    print(f"İndirimli tutar: {indirimli_toplam} TL")
```

```python
filmler = [
    {"ad": "Spirited Away", "yil": 2001, "puan": 8.6},
    {"ad": "Inception", "yil": 2010, "puan": 8.8},
    {"ad": "Parasite", "yil": 2019, "puan": 8.5},
]

def film_ekle(liste, ad, yil, puan):
    liste.append({"ad": ad, "yil": yil, "puan": puan})

arama = input("Film adı: ")
bulundu = False
for film in filmler:
    if film["ad"].lower() == arama.lower():
        print(f"Puan: {film['puan']}")
        film["puan"] = round(film["puan"] + 0.1, 2)
        bulundu = True
        break
if not bulundu: print("Film bulunamadı.")

yeni_filmler = [f for f in filmler if f["yil"] >= 2010]
print("2010 sonrası:", yeni_filmler)
```

```python
while True:
    try:
        veri = input("0-100 arası not (Çıkış: q): ")
        if veri.lower() == 'q': break
        n = float(veri)
        if 0 <= n <= 100:
            if n >= 90: h = "AA"
            elif n >= 80: h = "BA"
            elif n >= 70: h = "BB"
            elif n >= 60: h = "CB"
            elif n >= 50: h = "CC"
            else: h = "FF"
            print(f"Harf: {h}")
            break
        else: print("Geçersiz aralık!")
    except ValueError: print("Sayı giriniz!")
```

```python
import numpy as np

points = np.random.uniform(-5, 5, (200, 2))
distances = np.sqrt(np.sum(points**2, axis=1))
mask = distances <= 3

print(f"Seçilen: {np.sum(mask)}")
print(f"İçerdekiler Ort: {distances[mask].mean():.2f}")
print(f"Dışardakiler Ort: {distances[~mask].mean():.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(points[:, 0], points[:, 1], c='gray', alpha=0.3)
plt.scatter(points[mask, 0], points[mask, 1], c='blue')
plt.scatter(0, 0, c='red', marker='x', s=100)
plt.title("Nokta Dağılımı")

plt.subplot(1, 2, 2)
plt.hist(distances, bins=25, color='skyblue', edgecolor='black')
plt.axvline(3, color='red', linestyle='--')
plt.title("Mesafe Histogramı")

plt.tight_layout()
plt.show()
```
